```python
from pandas import DataFrame, read_csv
import pandas as pd 
import matplotlib.pyplot as plt
import datetime
import os
import numpy as np
```

```python
#FILL OUT NUMBERS OF BATTERIES WHICH HAVE BEEN TESTED: 
batteries = np.array([11, 12])
print("current directory " + os. getcwd() )
for x in batteries:
    dirName = '/data/batteryNum' + str(x)
    #print(dirName)
    os.listdir(os. getcwd() + dirName)
    print("go through the following file: " + dirName + '/' + os.listdir(os. getcwd() + dirName)[0])
    df = pd.read_csv(os. getcwd()+dirName + '/' + os.listdir(os. getcwd() + dirName)[0], sep = "\s+", header = None, skiprows = [0,1])
    df.loc[:, 2] = pd.to_datetime(df.loc[:, 2])
    df.loc[:, 2] = df.loc[:, 2] - df.loc[0, 2]
    fig, axs = plt.subplots(2)
    fig.suptitle('battery #' + str(x))
    axs[0].plot(df.loc[:, 2].astype('timedelta64[m]')/60,df.loc[:, 4],color="blue")
    axs[0].set_ylabel('voltage, [mV]')
    axs[1].plot(df.loc[:, 2].astype('timedelta64[m]')/60,df.loc[:, 5],color="red")
    axs[1].set_ylabel('current, [mA]')
    axs[1].set_xlabel('time from the start, hours')
    plt.savefig("pic/" + 'battery' + str(x)+'.pdf')
    print("pic/" + 'battery' + str(x)+'.pdf '+ 'has been created ')
```
